```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from collections import Counter
import warnings # Turn off warnings
warnings.filterwarnings('ignore')
```

```python
# Import
physics_df = pd.read_csv('data/physics-df.csv')
physics_df = physics_df.drop(columns = ["Unnamed: 0"])
physics_df = physics_df[['GPA', 'GRE_Verbal', 'GRE_Quantitative', 'GRE_Writing', 'GRE_Subject', 'American', 'Papers', 'Research', 'Decision']] # Remove unwanted columns
```

```python
# Everyone
physics_df = physics_df[['GPA', 'GRE_Verbal', 'GRE_Quantitative', 'GRE_Writing', 'GRE_Subject', 'Decision']]
sns.pairplot(physics_df, hue = "Decision")
plt.savefig("figures/pairplot.png", transparent=True, bbox_inches="tight")
```

```python
# Perfect GPA
sns.pairplot(physics_df[physics_df['GPA'] == 4.0], hue = "Decision")
plt.savefig("figures/pairplot-perfect-gpa.png", transparent=True, bbox_inches="tight")
```

```python
# Perfect GRE
sns.pairplot(physics_df[physics_df['GRE_Subject'] == 990], hue = "Decision")
plt.savefig("figures/pairplot-perfect-gre.png", transparent=True, bbox_inches="tight")
```

```python
# Perfect GPA and GRE
sns.pairplot(physics_df[physics_df['GPA'] == 4][physics_df['GRE_Subject'] == 990], hue = "Decision")
plt.savefig("figures/pairplot-perfect-gpa-and-gre.png", transparent=True, bbox_inches="tight")
```
